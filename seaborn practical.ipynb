# Multiple Plots (Facet-Grid)

```python
var = sns.load_dataset("tips")
var.head()
```

```python
sns.set_style("white")

sns.set_context("paper",font_scale=1.3)

fg=sns.FacetGrid(var,col="sex",hue="day",palette="CMRmap")

fg.map(plt.scatter,"total_bill","tip").add_legend()

plt.show()
```

```python
sns.set_context("paper",font_scale=1.3)

fg=sns.FacetGrid(var,col="day",hue="sex",palette="cool")

fg.map(plt.scatter,"total_bill","tip").add_legend()

plt.show()
```

```python
sns.set_context("paper",font_scale=1.3)

fg=sns.FacetGrid(var,col="sex",hue="day",palette="summer")

fg.map(plt.bar,"total_bill","tip",edgecolor="b").add_legend()

plt.show()
```
